### image complement

```python
import cv2   

img = cv2.imread('rn.jpg')
cv2.imshow('Original Image', img)
comp_img = 255 - img
cv2.imshow("Complemented Image", comp_img)
# bitwise_comp = cv2.bitwise_not(img)
# cv2.imshow('Bitwise complement', bitwise_comp)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### image thresholding

```python
import cv2 as cv
# import numpy as np
# from matplotlib import pyplot as plt

img = cv.imread('rn.jpg')
ret,thresh1 = cv.threshold(img,127,255,cv.THRESH_BINARY)
ret,thresh2 = cv.threshold(img,127,255,cv.THRESH_BINARY_INV)
ret,thresh3 = cv.threshold(img,127,255,cv.THRESH_TRUNC)
ret,thresh4 = cv.threshold(img,127,255,cv.THRESH_TOZERO)
ret,thresh5 = cv.threshold(img,127,255,cv.THRESH_TOZERO_INV)
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]
for i in range(6):
    cv.imshow(titles[i], images[i])    
cv.waitKey(0)
cv.destroyAllWindows()
```

### histogram calculation

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('rn.jpg', 0)
img_hist = cv2.calcHist([img], [0], None, [256], [0, 256])
plt.plot(img_hist)
plt.title('histogram')
```

### Histogram equilization

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('rn.jpg', 0)
equi_hist = cv2.equalizeHist(img)
cv2.imshow('Histogram Equilization', equi_hist)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
